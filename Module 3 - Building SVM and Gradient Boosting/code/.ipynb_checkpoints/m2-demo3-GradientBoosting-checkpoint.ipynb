# Gradient Boosting


#### Gradient Boosting

```python
import pandas as pd
print(pd.__version__)
```

```python
auto_data = pd.read_csv('../data/imports-85.data', sep=r'\s*,\s*', engine='python')
```

```python
auto_data
```

```python
import numpy as np

auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['price'].describe()
```

```python
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce')
```

```python
auto_data['price'].describe()
```

```python
auto_data = auto_data.drop('normalized-losses', axis=1)
```

```python
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['num-of-cylinders'].describe()
```

```python
cylinders_dict = { 'two': 2,
                  'three' : 3,
                  'four' : 4,
                  'five' : 5,
                  'six' : 6,
                  'eight' : 8,
                  'twelve' : 12    
}

auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)

auto_data.head()
```

```python
auto_data = pd.get_dummies(auto_data, columns=['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'fuel-system'])
```

```python
auto_data.head()
```

```python
auto_data = auto_data.dropna()
auto_data
```

```python
auto_data[auto_data.isnull().any(axis=1)]
```

```python
from sklearn.model_selection import train_test_split

X = auto_data.drop('price', axis=1)

# Taking the labels (price)
Y = auto_data['price']

# Spliting into 80% for training set and 20% for testing set so we can see our accuracy
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

params = {'n_estimators' : 500, 'max_depth' : 6, 'min_samples_split' : 2, 'learning_rate' : 0.01, 'loss' : 'ls'}

gbr_model = GradientBoostingRegressor(**params)

gbr_model.fit(X_train, Y_train)
```

```python
gbr_model.score(X_train, Y_train)
```

```python
y_predict = gbr_model.predict(x_test)
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15,6)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('MPG')

plt.legend()
plt.show()
```

```python
r_square = gbr_model.score(x_test, y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

gbr_model_mse = mean_squared_error(y_predict, y_test)
gbr_model_mse
```

```python
import math

math.sqrt(gbr_model_mse)
```

```python
from sklearn.model_selection import GridSearchCV

num_estimators = [100, 200, 500]
learn_rates = [0.01, 0.02, 0.05, 0.1]
max_depths = [4, 6, 8]

param_grid = {'n_estimators' : num_estimators,
              'learning_rate' : learn_rates,
              'max_depth' : max_depths}

grid_search = GridSearchCV(GradientBoostingRegressor(min_samples_split=2, loss='ls'),
                           param_grid, cv=3, return_train_score=True)

grid_search.fit(X_train, Y_train)

grid_search.best_params_
```

```python
grid_search.cv_results_
```

```python
for i in range(36):
    print('Parameters: ', grid_search.cv_results_['params'][i])
    print('Meant Test Score: ', grid_search.cv_results_['mean_test_score'][i])
    print('Rank: ', grid_search.cv_results_['rank_test_score'][i])
    print()
```

```python
params = {'n_estimators': 100, 'max_depth' : 4, 'min_samples_split' : 2, 'learning_rate' : 0.05, 'loss' : 'ls'}
gbr_model = GradientBoostingRegressor(**params)
gbr_model.fit(X_train, Y_train)
```

```python
y_predict = gbr_model.predict(x_test)
```
